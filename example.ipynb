```python
from curved_slits import PointPicker, spline_slit

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Here is quick example of how to use generate perpendicular slits from a guide.

We use here an example image with a known functional form for the intensities in order to demonstrate the accuracy.

First, let us create the fake image.

```python
x = np.linspace(0,10,100)
X,Y = np.meshgrid(x,x)
z = np.sin(X)**2 + np.log(Y+1)

plt.imshow(z, origin='lower')
```

We now want to select a number of points that will be used to create a spline line, which will become our guide line. From the guide we will calculate the perpendicular slits.

First, we need to select our points interactively. This doesn't work in the notebook, but I show the result of running the line below.

```python
out = PointPicker(z)
```

In the next cell we re-run the above line directing the output to an operating system window, with which one can be interactive. Note that this does not need to be done if running from the python command line. 

In the window you can select a number of points interactively to define your guide line. There is no limit to the number of points, but they should be in order. 

The x-y coordinates of the selected points are printed to the screen.

```python
%matplotlib osx
out = PointPicker(z)
```

Now I just show the points I selected (the red crosses appear on the interactive screen after a point has been selected).

```python
%matplotlib inline
points = out.return_points()

plt.imshow(z, origin='lower')
plt.plot(points[0], points[1], '+', color='red')
```

We now pass the points to the `spline_slit` routine, which will generate the spline guide line and calculate the slits perpendicular to the spline line. The pixel separation between slits can be changed (default `spacing=1`) and the accuracy of the separation distance (default `diff_tol=1e-2`) can be specified by keywords. In order to get accurate slit separations, the oversample keyword needs to set to `True`.


Note that we add an additional axis to our image, as the routine expects a 3D cube of images - with the frame number along the zero axis.

```python
print(z[np.newaxis,].shape)
res = spline_slit(z[np.newaxis,], points, plot_slit=True, oversample=True)
```

The above figure shows the selected guide points, the spline line and the slits where data is extracted from the image.

To check the accuracy of the interpolation that takes place when extracting the data, we show the true intensity values against the values along the guide line.

```python
slit = res[0]
spline_points = res[1][0]
z_true = np.sin(spline_points[0]*x[1])**2+np.log(spline_points[1]*x[1]+1)

# guideline located at the center of the slit
plt.plot(slit[:,21//2])
plt.plot(z_true, '+')
```

Plot of differences.

```python
diff = slit[:,21//2,0]-z_true
print('Max difference', np.abs(diff).max())
plt.plot(diff)
```
